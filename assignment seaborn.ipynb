Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot

```python
# 1.scatterplot()
# 2.displot()
# 3.relplot()
# 4.catplot()
# 5.joinplot()
```

```python
import seaborn as sns
# 1.scatterplot()

iris=sns.load_dataset("iris")
iris.head()
```

```python
sns.scatterplot(x=iris.sepal_length,y=iris.sepal_width)
```

```python
# 2.displot()
sns.displot(iris['petal_width'])
```

```python
# 3.relplot()
sns.relplot(x=iris.sepal_width,y=iris.sepal_length,data=iris,hue='species')
```

```python
# 4.catplot()
sns.catplot(x=iris.petal_length,y=iris.species)
```

```python
# 5.joinplot()
sns.jointplot(x=iris.sepal_length,y=iris.sepal_width)
```
