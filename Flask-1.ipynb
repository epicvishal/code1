Question 5. What function is used in Flask for URL Building? Write a python code to demonstrate the working of the url_for()function. 

```python
# The url_for() function is used in Flask for URl building. It takes the name of the fuction or view as its argument and
# any additional keyword arguments of the dynamic parts of the URl as its subsequent arguments.It then generates a URL 
# for the specified function or view.




from flask import Flask

app = Flask(__name__)

@app.route("/welcome")
def hello_world():
    return "<h1> Welcome to ABC Corporation</h1>"


@app.route("/")
def details():
    return "<h1>Company Name:ABC Corporation  Location:India  Contact Details:999-999-9999</h1>"

if __name__=="__main__"
    app.run(host="0.0.0.0")
```
