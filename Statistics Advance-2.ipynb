Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with
an example.

```python
#The Probability Mass Function (PMF) and Probability Density Function (PDF) are both mathematical functions used to describe the probability distribution of a random variable.

#1. Probability Mass Function (PMF):
#The PMF is used for discrete random variables. 
#It gives the probability that a discrete random variable takes on a specific value. 
#In other words, it maps each possible value of the random variable to its probability.

import numpy as np

def fair_die_pmf(x):
    if x in [1,2,3,4,5,6]:
        return 1/6
    else:
        return 0
    

x_values = np.arange(1,7)
pmf_values = [fair_die_pmf(x) for x in x_values]

print("PMF values:",pmf_values)
```

```python
#2. Probability Density Function (PDF):
#The PDF is used for continuous random variables. 
#It gives the probability density at each point in the range of possible values of the random variable. 
#The area under the PDF curve over a certain interval represents the probability of the random variable falling within that interval.

import numpy as np
from scipy.stats import norm 
import matplotlib.pyplot as plt

samples = np.random.randn(1000)

pdf_values = norm.pdf(samples)

plt.plot(samples, pdf_values)
plt.xlabel('x')
plt.ylabel('PDF')
plt.title("standard Normal Distribution PDF")
plt.show()
```

Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?

```python
#The Cumulative Density Function (CDF) is a mathematical function that gives the probability that a random variable takes on a value less than or equal to a given value.
#In other words, it calculates the cumulative probability up to a certain point.

#The CDF is used to understand the probability distribution of a random variable and to calculate probabilities of events occurring within a certain range.

import numpy as np
from scipy.stats import norm

x = 1.5
cdf_value = norm.cdf(x)
print("CDF value:",cdf_value)
```

```python
#The CDF is used for various purposes, including:

#1. Calculating probabilities: The CDF can be used to calculate the probability of a random variable falling within a certain range. 
#By subtracting the CDF values at two different points, we can find the probability of the random variable falling between those two points.

#2. Comparing distributions: The CDF allows us to compare different probability distributions by visualizing their cumulative probabilities. 
#This helps in understanding the shape and characteristics of different distributions.

#3. Generating random samples: The inverse of the CDF, known as the Quantile Function or Percent-Point Function, can be used to generate random samples from a given distribution.
#By generating random numbers from a uniform distribution and applying the inverse CDF, we can obtain samples from the desired distribution.
```

Q3: What are some examples of situations where the normal distribution might be used as a model?
Explain how the parameters of the normal distribution relate to the shape of the distribution.

```python
#The normal distribution, also known as the Gaussian distribution or bell curve, is widely used as a model in various fields due to its simplicity and applicability to many real-world situations. Some examples of situations where the normal distribution might be used as a model include:

#1. Height: The heights of individuals in a population tend to follow a roughly normal distribution, with most people being of average height and fewer people being either very tall or very short.

#2. IQ Scores: Intelligence quotient (IQ) scores are often assumed to follow a normal distribution, with the majority of people having average intelligence and fewer people having either high or low intelligence.

#3. Errors in Measurements: When measuring physical quantities, there is often some degree of error involved. These errors can be modeled using a normal distribution, assuming that they are random and follow a bell-shaped pattern.

#4. Financial Returns: In finance, it is often assumed that the returns on investments or stock prices follow a normal distribution. This assumption allows for the use of various statistical models and techniques for risk analysis and portfolio management.

#The shape of the normal distribution is determined by its two parameters: the mean (μ) and the standard deviation (σ). The mean represents the center or average value of the distribution, while the standard deviation measures the spread or variability of the data around the mean.

#1. Mean (μ): The mean determines the location of the peak or center of the normal distribution. It represents the average value around which the data is symmetrically distributed. Shifting the mean to the right or left will shift the entire distribution along the x-axis.

#2. Standard Deviation (σ): The standard deviation controls the spread or dispersion of the data. A smaller standard deviation indicates that the data points are closely clustered around the mean, resulting in a narrower and taller bell curve. Conversely, a larger standard deviation leads to a wider and flatter distribution.
```

```python
#The normal distribution is important for several reasons:

#1. Central Limit Theorem: The normal distribution plays a crucial role in the Central Limit Theorem, which states that the sum or average of a large number of independent and identically distributed random variables will be approximately normally distributed, regardless of the shape of the original distribution. This theorem is fundamental in statistics and allows for the use of various statistical techniques and hypothesis tests.

#2. Inference and Estimation: Many statistical inference and estimation methods are based on the assumption of normality. For example, in hypothesis testing, we often assume that the test statistic follows a normal distribution under the null hypothesis. Similarly, in regression analysis, we assume that the residuals follow a normal distribution.

#3. Confidence Intervals: Normal distributions are used to construct confidence intervals, which provide a range of values within which an unknown parameter is likely to fall. The construction of confidence intervals relies on the assumption of normality.

#4. Decision Making: Normal distributions are often used as models for decision-making processes. For example, in quality control, if a process is normally distributed, it is easier to set appropriate control limits and make decisions about whether the process is in control or out of control.

#Real-life examples of situations where the normal distribution is commonly observed include:

#1. Exam Scores: In education, exam scores often follow a normal distribution. Most students score around the average, with fewer students scoring very high or very low.

#2. Body Measurements: Various body measurements such as weight, height, and blood pressure tend to follow a normal distribution in a population.

#3. Product Quality: The quality characteristics of manufactured products, such as length, width, or weight, are often assumed to be normally distributed.

#4. Stock Market Returns: The daily returns of stocks or stock market indices are often assumed to be normally distributed. This assumption allows for the use of various risk models and portfolio optimization techniques.
```

Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli
Distribution and Binomial Distribution?

```python
#The Bernoulli distribution is a discrete probability distribution that models a single trial with two possible outcomes: success (usually represented as 1) or failure (usually represented as 0). The distribution is characterized by a single parameter, p, which represents the probability of success.

#An example of the Bernoulli distribution is flipping a fair coin. The outcome of each flip can be either heads (success) or tails (failure), and the probability of getting heads is 0.5.

#The binomial distribution, on the other hand, models the number of successes in a fixed number of independent Bernoulli trials. It is characterized by two parameters: n, the number of trials, and p, the probability of success in each trial.

#The key difference between the Bernoulli distribution and the binomial distribution is that the Bernoulli distribution models a single trial, while the binomial distribution models multiple trials. The binomial distribution provides the probability of getting a specific number of successes in a fixed number of trials.

#For example, suppose we flip a fair coin 5 times and want to know the probability of getting exactly 3 heads. This can be modeled using the binomial distribution with n = 5 (number of trials) and p = 0.5 (probability of success in each trial).
```

Q6. Consider a dataset with a mean of 50 and a standard deviation of 10. If we assume that the dataset
is normally distributed, what is the probability that a randomly selected observation will be greater
than 60? Use the appropriate formula and show your calculations.

```python
import scipy.stats as stats

mean = 50
std_dev = 10
x = 60

z_score = (x -mean)/ std_dev
probability = 1-stats.norm.cdf(z_score)

print("Probability:",probability)
```

Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.

```python
#The Central Limit Theorem (CLT) is a fundamental concept in statistics that states that the sampling distribution of the sample mean of any independent, identically distributed (i.i.d.) random variables will be approximately normally distributed, regardless of the original distribution of the population, as long as the sample size is sufficiently large.

#In other words, if you take repeated samples from a population and calculate the mean of each sample, the distribution of these sample means will approach a normal distribution as the sample size increases, regardless of the shape of the population distribution.

#Key points about the significance of the Central Limit Theorem include:

#1. Robustness: The CLT is incredibly robust and applies to a wide range of distributions, including those that are not normal. This allows statisticians to use normal distribution properties for inference, even when dealing with non-normally distributed populations.

#2. Foundation of Statistical Inference: The CLT is the foundation for many statistical inference techniques, such as hypothesis testing and confidence interval estimation. It enables us to make inferences about population parameters based on sample statistics.

#3. Large Sample Sizes: The CLT becomes increasingly accurate as the sample size increases. This means that even if the population distribution is unknown or non-normal, we can rely on the normality of the sample means for making statistical inferences with large sample sizes.

#4. Practical Applications: The CLT has numerous practical applications across various fields, including finance, economics, biology, psychology, and quality control. It is used for estimating population parameters, conducting hypothesis tests, and constructing confidence intervals.

#5. Sampling Strategies: Understanding the CLT helps researchers and practitioners design sampling strategies and determine appropriate sample sizes for their studies. It provides guidelines for ensuring the validity and reliability of statistical analyses.
```
